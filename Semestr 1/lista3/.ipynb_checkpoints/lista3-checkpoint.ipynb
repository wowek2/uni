# Zadanie 1 i 2
Odczyt danych z pliku tekstowego (Dwie pierwsze strony Harrego Pottera):

```python
text_file_en = open('tekst_lista3_en.txt', encoding='utf-8',mode = 'r')
text_file_pl = open('tekst_lista3_pl.txt', encoding='utf-8',mode = 'r')
content_en = text_file_en.read()
content_pl = text_file_pl.read()
```

Zapis liczby wystąpień poszczególny liter do słownika:

```python
def count_letters(content):
    letters_count = {}
    for letter in content:
        if letter.isalpha():
            if letter.upper() in letters_count.keys():
                letters_count[letter.upper()] += 1
            else:
                letters_count[letter.upper()] = 1
    return letters_count

letters_count_en = count_letters(content_en)
letters_count_pl = count_letters(content_pl)
print(letters_count_pl)
print(letters_count_en)
```

Przedstawienie histogramu 10 najczęściej występujących liter w języku polskim oraz agnielskim:

```python
top_10_en = dict(sorted(letters_count_en.items(), key = lambda x: x[1], reverse = True)[:10])
top_10_pl = dict(sorted(letters_count_pl.items(), key = lambda x: x[1], reverse = True)[:10])

plt.subplot(2,1,1)
plt.bar(top_10_en.keys(), top_10_en.values(), color="green")

plt.xlabel("Litera")
plt.ylabel("Liczba wystąpień")
plt.title("Liczba wystąpień 10 najpopularniejszych liter (Język angielski)")

plt.subplot(2,1,2)
plt.bar(top_10_pl.keys(), top_10_pl.values(), color="green")

plt.xlabel("Litera")
plt.ylabel("Liczba wystąpień")
plt.title("Liczba wystąpień 10 najpopularniejszych liter (Język polski)")
plt.tight_layout()
plt.show()
```

```python
str = input("Podaj stringa do zakodowania szyfrem ROT13: ")
```

Rozwiązanie #1 z użyciem maping table: (nie obsługuje polskich znaków, przepisuje je)

```python
rot13 = bytes.maketrans(b"abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ",
    b"nopqrstuvwxyzabcdefghijklmNOPQRSTUVWXYZABCDEFGHIJKLM")
print(str.translate(rot13))
```

Rozwiązanie #2 z użyciem indeksowania ASCII: (nie obsługuje polskich znaków, przepisuje je)

```python
def rot13_1(str):
    return ''.join([chr((ord(letter) - 97 + 13) % 26 + 97)
                        if 97 <= ord(letter) <= 122 # Jeżeli litera pomiędzy a-z zamień o 13
                        else letter                 # Jeżeli litera z poza przedziału (a,z) zwróć tą samą literę (np. white space, !, ?)
                        for letter in str.lower()]) # string musi być z małych liter, bo małe i duże litery mają różne kodowanie w ASCII
print(rot13_1(str))                                 # ewentualnie można rozpisać funkcję i sprawdzać czy duża czy mała litera i odpowiednio kodować
```

Rozwiązanie #3 polski alfabet z użyciem mapping table:

```python
def rot13_2(str):
    mapping_table = str.maketrans( 
        'aąbcćdeęfghijklłmnńoópqrsśtuvwxyzźżAĄBCĆDEĘFGHIJKLŁMNŃOÓPQRSŚTUVWXYZŹŻ', 
        'klłmnńoópqrsśtuvwxyzźżaąbcćdeęfghijKLŁMNŃOÓPQRSŚTUVWXYZŹŻAĄBCĆDEĘFGHIJ'
    )
    return str.translate(mapping_table)
print(rot13_2(str))
```

# Zadanie 5

```python
def gcd(a,b):
    while b:
        a, b = b, a%b
    return a
```

```python
x, y = input("Podaj dwie liczby, aby znaleźć ich NWD:").split()
print('NWD: ', gcd(int(x),int(y)))
```

# Zadanie 6

```python
levels = input("Podaj liczbę wierszy, z której ma być zbudowany trójkąt Pascala: ")

def triangle(levels):
    if levels == 0:
        return []
    elif levels == 1:
        return [[1]]
    else:
        new_row = [1]
        result = triangle(levels-1)
        last_row = result[-1]
        for i in range(len(last_row)-1):
            new_row.append(last_row[i] + last_row[i+1])
        new_row += [1]
        result.append(new_row)
    return result

for i in triangle(int(levels)):
    print(i)
```
